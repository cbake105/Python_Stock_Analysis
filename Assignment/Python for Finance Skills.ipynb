```python
sp500oc['Simple_Return_SP500'] = (sp500oc['Close'] - sp500oc['Previous_Close_SP500']) / sp500oc['Previous_Close_SP500']
baoc['Simple_Return_BA'] = (baoc['Close'] - baoc['Previous_Close_BA']) / baoc['Previous_Close_BA']
```

```python
baoc.head()
```

```python
# Merge the two DataFrames
merged_data = pd.merge(sp500oc, baoc, left_index=True, right_index=True, suffixes=('_SP500', '_BA'))
```

```python
merged_data.head()
```

```python
# Standard Deviation of the Log and Simple returns
print("\nStandard Deviation of Simple Returns:\n", merged_data[['Simple_Return_SP500', 'Simple_Return_BA']].std())
```

# Exercise 5 - Rename columns and create lagged and moving average variables

```python
# Import CSV file
sp500_df = pd.read_csv('sp500.csv')
```

```python
# Convert columns to all lowercase and replace spaces with '_'
sp500_df.columns = [col.lower().replace(' ', '_') for col in sp500_df.columns]
```

```python
sp500_df.head()
```

### One measure of volatility is the weighted average of the squared return of the previous day close to current day open, and the squared return of the current day open to the current day close

#### Create variables for volatility measures and log returns

```python
# Add 'previous_close' column
sp500_df['previous_close'] = sp500_df['close'].shift(1)
```

```python
# Calculate log return of the previous day close to current day open
sp500_df['log_rtn_c_to_o'] = np.log(sp500_df['open'] / sp500_df['previous_close'])
```

```python
# Calculate log return of the current day open to the current day close
sp500_df['log_rtn_o_to_c'] = np.log(sp500_df['close'] / sp500_df['open'])
```

```python
# Create variables that square volatility measures
sp500_df['vol_c_to_o'] = sp500_df['log_rtn_c_to_o']**2
sp500_df['vol_o_to_c'] = sp500_df['log_rtn_o_to_c']**2
```

```python
# Create a variable representing weighted volatility
sp500_df['vol'] = 0.5 * (sp500_df['vol_c_to_o'] + sp500_df['vol_o_to_c'])
```

```python
# Create a lagged variable of 'vol'
sp500_df['vol_1'] = sp500_df['vol'].shift(1)
```

```python
# Create moving averages of 'vol'
sp500_df['vol_ma_5'] = sp500_df['vol'].rolling(window=5).mean()
sp500_df['vol_ma_22'] = sp500_df['vol'].rolling(window=22).mean()
```

```python
# Print the first 30 records
print(sp500_df.head(30))
```

# Exercise 6 - Merge data of different frequencies and create variables using financial report data

```python
# Import Boeing stock CSV file 
boeing_df = pd.read_csv('BA.csv')
```

```python
# Convert columns to all lowercase and replace spaces with '_'
boeing_df.columns = [col.lower().replace(' ', '_') for col in boeing_df.columns]
```

```python
# Convert 'Date' column to DateTime format 
boeing_df['date'] = pd.to_datetime(boeing_df['date'])
```

```python
# Import Boeing stock CSV file 
financials_df = pd.read_csv('fundamentals.csv')
```

```python
# Convert columns to all lowercase and replace spaces with '_'
financials_df.columns = [col.lower().replace(' ', '_') for col in financials_df.columns]
```

```python
# Convert 'Date' column to DateTime format
financials_df['period_ending'] = pd.to_datetime(financials_df['period_ending'])
```

```python
# Create variable named 'match_date' in stock and financial data
boeing_df['match_date'] = boeing_df['date'] - pd.DateOffset(months=3)
financials_df['match_date'] = financials_df['period_ending'] + pd.DateOffset(months=3)
```

```python
# Create variable named 'roa' in the financials DataFrame
financials_df['roa'] = financials_df['net_income'] / financials_df['total_assets']
```

```python
# Left merge the DataFrames
boeing_financials_df = pd.merge(boeing_df, financials_df[['match_date', 'roa']],left_on='match_date', right_on='match_date', how='left')
```

```python
# Fill in the NAs
boeing_financials_df.fillna(method='ffill', inplace=True)
```

```python
# Create variables from financial DataFrame
financials_df['max_date'] = financials_df.groupby('ticker_symbol')['match_date'].transform('max')
financials_df['min_date'] = financials_df.groupby('ticker_symbol')['match_date'].transform('min')
```

```python
ba_max_date = financials_df.loc[financials_df['ticker_symbol'] == 'BA', 'max_date'].max()
ba_min_date = financials_df.loc[financials_df['ticker_symbol'] == 'BA', 'min_date'].min()
```

```python
# Reassign the merged DataFrame
boeing_financials_df = boeing_financials_df[(boeing_financials_df['date'] <= ba_max_date) & (boeing_financials_df['date'] >= ba_min_date)]
```

```python
print(boeing_financials_df['roa'].info())
```

```python
# Drop NaN's from 'roa' column
boeing_financials_df = boeing_financials_df.dropna(subset=['roa'])
```
